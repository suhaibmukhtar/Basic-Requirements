```python
def display_options():
    print("1.Sinup\n2.Signin\n3.Quit")

def validate(name,address,contact,password,c_password,dob):
    if not name or not contact or not password or not c_password or not dob:
        print('Field values should not be empty!')
        return -1
    #check whether name starts with 0 and is of 10 digits
    pattern='^0[0-9]{9}'
    match_contact=re.compile(pattern)
    res1=match_contact.search(contact)
    if not res1:
        print('Please enter the contact in correct format!')
        return -1
    ##Chech whether password contains alphabet(@&$)numeric
    pattern2='^[a-zA-Z]+[@$&][0-9]+'
    match_pass=re.compile(pattern2)
    res2=match_pass.search(password)
    if not res2:
        print('Please enter the email in valid format!')
        return -1
    #whether password matching
    if password!=c_password:
        print("Passwords are not matching!")
        return -1
    #check the dob format 15/07/2001 or 15-07-2001
    pattern3="^\d{2}[-\/]\d{2}[-\/]\d{4}"
    match_dob=re.compile(pattern3)
    res3=match_dob.search(dob)
    if not res3:
        print("Please enter Dob in Valid format")
        return -1
    current_year=current_year = datetime.now().year
    year = re.search(r'\d{4}', dob).group()
    val=current_year-int(year)
    if val<21:
        print("Age should be greater than 21")
        return -1
    return 1


def main():
    users={}
    check=0
    while True:
        display_options()
        ch=int(input('Enter your choice:'))
        if ch==1:
            name=input('Enter name:')
            address=input('Enter address or skip press enter:')
            contact=input('Enter phone number:')
            password=input('Enter password:')
            c_password=input('Enter the Passoword:')
            dob=input("Enter dob in format #DD-MM-YYYY:")
            check=validate(name,address,contact,password,c_password,dob)
            if check==-1:
                print('Signup Failed!')
                continue
            elif check==1:
                user=User(name,address,contact,password,dob)
                user.signup()
        elif ch==2:
            if check==0:
                print("Please Signup First!")
                continue
            else:
                k=user.signin()
                if k==-1:
                    continue
                else:
                    print("Login Successfull!")
                    c=user.menu()
                    ##Divine
                    if c==1:
                        user.new_menu()
                    
        elif ch==3:
            exit(0)
        else:
            print("please enter valid choice!")

if __name__=='__main__':
    main()
```
