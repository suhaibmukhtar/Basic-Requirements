```python
import pandas as pd

class OrderProcessor:
    def __init__(self):
        # Define an empty DataFrame in the constructor
        self.order_df = pd.DataFrame(columns=['ID', 'Total Amount', 'Order Type'])

    def delivery(self):
        while True:
            try:
                order_id = int(input("Enter order ID: "))
                total_amount = float(input("Enter total amount: "))
                # Create a dictionary for delivery orders
                order_dict = {'ID': order_id, 'Total Amount': total_amount, 'Order Type': 'Delivery'}
                # Append the dictionary to the DataFrame
                self.order_df = self.order_df.append(order_dict, ignore_index=True)
                break  # Exit the loop if inputs are valid
            except ValueError:
                print("Invalid input. Please enter a valid order ID and total amount.")

    def dine(self):
        while True:
            try:
                order_id = int(input("Enter order ID: "))
                total_amount = float(input("Enter total amount: "))
                # Create a dictionary for dine-in orders
                order_dict = {'ID': order_id, 'Total Amount': total_amount, 'Order Type': 'Dine-In'}
                # Append the dictionary to the DataFrame
                self.order_df = self.order_df.append(order_dict, ignore_index=True)
                break  # Exit the loop if inputs are valid
            except ValueError:
                print("Invalid input. Please enter a valid order ID and total amount.")

    def selfpick(self):
        while True:
            try:
                order_id = int(input("Enter order ID: "))
                total_amount = float(input("Enter total amount: "))
                # Create a dictionary for self-pickup orders
                order_dict = {'ID': order_id, 'Total Amount': total_amount, 'Order Type': 'Self-Pickup'}
                # Append the dictionary to the DataFrame
                self.order_df = self.order_df.append(order_dict, ignore_index=True)
                break  # Exit the loop if inputs are valid
            except ValueError:
                print("Invalid input. Please enter a valid order ID and total amount.")

# Example usage
order_processor = OrderProcessor()
while True:
    print("1. Delivery\n2. Dine-In\n3. Self-Pickup\n4. Exit")
    choice = input("Enter your choice: ")
    if choice == '1':
        order_processor.delivery()
    elif choice == '2':
        order_processor.dine()
    elif choice == '3':
        order_processor.selfpick()
    elif choice == '4':
        break
    else:
        print("Invalid choice. Please select again.")
```
